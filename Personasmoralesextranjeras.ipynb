```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler, normalize
from google.colab import drive
```

```python
# Contar la cantidad de productores por estado
estado_counts = extrenjeras_df['ESTADO'].value_counts().reset_index()
estado_counts.columns = ['ESTADO', 'Cantidad']

# Graficar la cantidad de productores por estado
plt.figure(figsize=(12, 6))
sns.barplot(x='ESTADO', y='Cantidad', data=estado_counts)
plt.title('Cantidad de Productores por Estado')
plt.xticks(rotation=90)
plt.show()
```

```python
# Contar la cantidad de productores por municipio
municipio_counts = extrenjeras_df['MUNICIPIO'].value_counts().reset_index()
municipio_counts.columns = ['MUNICIPIO', 'Cantidad']

# Graficar la cantidad de productores por municipio
plt.figure(figsize=(20, 6))
sns.barplot(x='MUNICIPIO', y='Cantidad', data=municipio_counts)
plt.title('Cantidad de Productores por Municipio')
plt.xticks(rotation=90)
plt.show()
```

```python
# eliminacion de municiopio
extrenjeras_df.drop(columns=['MUNICIPIO',], inplace=True)
```

```python
# Contar la cantidad de productores por país de origen
pais_counts = extrenjeras_df['PAÍS DE ORIGEN'].value_counts().reset_index()
pais_counts.columns = ['PAÍS DE ORIGEN', 'Cantidad']

# Graficar la cantidad de productores por país de origen
plt.figure(figsize=(12, 6))
sns.barplot(x='PAÍS DE ORIGEN', y='Cantidad', data=pais_counts)
plt.title('Cantidad de Productores por País de Origen')
plt.xticks(rotation=90)
plt.show()
```

```python
# Contar la cantidad de productores por clave SCIAN 2023
scian_counts = extrenjeras_df['CLAVE SCIAN 2023'].value_counts().reset_index()
scian_counts.columns = ['CLAVE SCIAN 2023', 'Cantidad']

# Graficar la cantidad de productores por clave SCIAN 2023
plt.figure(figsize=(12, 6))
sns.barplot(x='CLAVE SCIAN 2023', y='Cantidad', data=scian_counts)
plt.title('Cantidad de Productores por Clave SCIAN 2023')
plt.xticks(rotation=90)
plt.show()
```

```python
# eliminacion de id
extrenjeras_df.drop(columns=['CLAVE SCIAN 2023',], inplace=True)
```

```python
extrenjeras_df['FECHA INICIO OPERACIONES'] = extrenjeras_df['FECHA INICIO OPERACIONES'].astype(str)
extrenjeras_df['FECHA INICIO OPERACIONES'] = extrenjeras_df['FECHA INICIO OPERACIONES'].str[-4:]
```

```python
# Contar la cantidad de productores por fecha de inicio de operaciones
fecha_counts = extrenjeras_df['FECHA INICIO OPERACIONES'].value_counts().reset_index()
fecha_counts.columns = ['FECHA INICIO OPERACIONES', 'Cantidad']

# Graficar la cantidad de productores por fecha de inicio de operaciones
plt.figure(figsize=(12, 6))
sns.barplot(x='FECHA INICIO OPERACIONES', y='Cantidad', data=fecha_counts)
plt.title('Cantidad de Productores por Fecha de Inicio de Operaciones')
plt.xticks(rotation=90)
plt.show()
```
